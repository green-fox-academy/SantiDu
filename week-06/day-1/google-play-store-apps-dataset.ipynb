Which app not to download? - Find an app which takes the most space on your phone, but is rated poorly (define poorly)?

```python
def number(x):
    if x[-1] == "M":
        return float(x[:-1]) * 1024
    elif x[-1] == "k":
        return float(x[:-1])
    elif x == "Varies with device":
        return None

df["Size in k"] = df["Size"].apply(number)
```

```python
largeapps = df.sort_values(by="Size in k", ascending=False)
```

```python
largeapps[largeapps["Rating"] < 3][:10]
```
