```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Superstore Sales (Excel).xls')
```

```python
df.info()
```

```python
df["Order Priority"].unique()
```

```python
df[(df["Ship Date"].dt.year == 2010) & (df["Order Priority"] == "High")].groupby("Ship Date")["Row ID"].count().plot(figsize=(10,5))
```

```python
df.groupby("Product Container")["Shipping Cost"].agg([max, min]).plot.bar(figsize=(10, 5))
```

```python
df.groupby(["Product Category", "Region"])["Profit"].sum().unstack().plot.bar(figsize=(10, 5))
```
